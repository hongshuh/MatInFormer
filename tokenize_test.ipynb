```python
laue_classes = {
        '-1': ['1','-1'],
        '2/m':['2','m','2/m'],
        'mmm':['222','mm2','mmm'],
        '4/m':['4','-4','4/m'],
        '4/mmm':['422','4mm','-42m','-4m2','4/mmm'],
        '-3':['3','-3'],
        '-3m':['32','321','312','3m1','31m','3m','-3m','-3m1','-31m'],
        '6/m':['6','-6','6/m'],
        '6/mmm':['622','6mm','-6m2','-62m','6/mmm'],
        'm-3':['23','m-3'],
        'm-3m':['432','-43m','m-3m']
        }
symmetry_class = {
        'Centrosymmetric':['-1','2/m','mmm','4/m','4/mmm','-3m','-31m','-3m1','-3','6/m','6/mmm','m-3','m-3m'],
        'Enantiomorphic':['1','2','222','4','422','3','32','312','321','6','622','23','432'],
        'None':['m','mm2','-4','4mm','-42m','-4m2','3m','31m','3m1','-6','6mm','-6m2','-62m','-43m']
    }
polar_calss = ['1','2','m','mm2','4','4mm','3','3m','3m1','31m','6','6mm']
def token_spacegroup(full_symbol):
    tokens = []
    pattern = r'([A-Z]|[0-9]|[-_/]|[a-z])'
    seperate_str = re.findall(pattern,full_symbol)
    i = 0

    while i < len(seperate_str):
        if seperate_str[i] in ['_','/']:
            tokens[-1] = tokens[-1]+seperate_str[i] + seperate_str[i+1]
            i+=2
        elif seperate_str[i] == '-':
            token = seperate_str[i] + seperate_str[i+1]
            tokens.append(token)
            i+=2
        else:
            token = seperate_str[i]
            tokens.append(token)
            i+=1
    while len(tokens) < 4:
        tokens.append('1')
    return tokens

with open('/home/hongshuh/space_group_transformer/sgt/embeddings/wyckoff-position-multiplicities.json') as file:
    wyckoff_multiplicity_dict = json.load(file)
with open('/home/hongshuh/space_group_transformer/sgt/embeddings/wyckoff-position-params.json') as file:
    param_dict = json.load(file)

def token_wkf(spg_num):
    wkf_tokens = []
    return wkf_tokens
```

```python
all_spg = {}
for keys in pg.keys():
    spg_dict = {}
    spg = SpaceGroup(keys)
    full_symbol = pg[keys]['full_symbol']
    spg_dict['full_symbol'] = pg[keys]['full_symbol']
    spg_dict['spg_number'] = 'spg_'+str(pg[keys]['int_number'])
    spg_dict['order'] = str(float(pg[keys]['order']))
    spg_dict['point_group'] = pg[keys]['point_group']
    spg_dict['crystal_system'] = spg.crystal_system

    for key, values in laue_classes.items():
        if spg_dict['point_group'] in values:
            spg_dict['laue_class'] = key
    
    for key, values in symmetry_class.items():
        if spg_dict['point_group'] in values:
            spg_dict['symmetry_class'] = key

    if spg_dict in polar_calss:
        spg_dict['polar'] = 'polar'
    else:
        spg_dict['polar'] = 'non_polar'

    tokens = token_spacegroup(spg_dict['full_symbol'])
    for i in range(4):
        spg_dict[f'spg_token_{i}'] = tokens[i]
    
    all_spg[full_symbol] = spg_dict
all_spg['P12_1/m1']
```

```python
def add_values_to_set(value, unique_values):
    if isinstance(value, list):
        for item in value:
            add_values_to_set(item, unique_values)
    elif isinstance(value, dict):
        for item in value.values():
            add_values_to_set(item, unique_values)
    else:
        unique_values.add(value)

def unique_values_from_dicts(dict1, dict2):
    unique_values = set()
    
    for value in dict1.values():
        add_values_to_set(value, unique_values)
        
    for value in dict2.values():
        add_values_to_set(value, unique_values)
    
    return list(unique_values)
unique_values_list = unique_values_from_dicts(all_wkf_tokens, all_spg)
len(unique_values_list)
```

```python
def tokenize_space_group(data):
    '''
    
    '''
    tokens = []
    structure = Structure.from_dict(data)

    comp = structure.composition
    elem_dict = comp.get_el_amt_dict()
    for e,n in elem_dict.items():
        for i in range(int(n)):
            tokens.append(e)

    space_group_str,space_group_num = structure.get_space_group_info()
    spacegroup = SpaceGroup(space_group_str)
    pattern = r'([A-Z]+|[0-9]+|[-_/]|[a-z])'
    seperate_str = re.findall(pattern,space_group_str)

    i = 0
    seperate_space_group = []
    while i < len(seperate_str):
        idx = i
        cur_str = seperate_str[idx]
        if cur_str in ['/','_']:
            seperate_space_group[idx-1] = seperate_str[idx-1] + cur_str + seperate_str[idx+1]
            i += 2
        elif cur_str == '-':
            token = cur_str+seperate_str[idx+1]
            seperate_space_group.append(token)
            i += 2
        else:
            token = seperate_str[idx]
            seperate_space_group.append(token)
            i+=1
    
    crystal_system = spacegroup.crystal_system
    point_group = spacegroup.point_group
    order = spacegroup.order
    print(point_group)
    for lc, pgs in laue_classes.items():
        # print(lc)
        if point_group in pgs:
            laue_class = lc
            break
    for sc,pgs in symmetry_class.items():
        if point_group in pgs:
            symmetry = sc
            break

    tokens += seperate_space_group
    tokens.append(symmetry)
    tokens.append(crystal_system)
    tokens.append(laue_class)
    tokens.append(point_group)
    tokens.append(order)
    tokens.append(space_group_num)
    return tokens
```
